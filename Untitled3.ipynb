```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(-5.0,5.0,0.1)
y = np.maximum(0,x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y=x
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
x = np.arange(-5.0,5.0,0.1)
exp_x = np.exp(x)
sum_exp_x = np.sum(exp_x)
y = exp_x / sum_exp_x
plt.plot(x,y)
plt.show()
```
